**FEATURES SELECTION**

```python
df_features=df[['Movie_Genre','Movie_Language','Movie_Popularity','Movie_Keywords','Movie_Cast']].fillna('')
```

```python
df_features['Movie_Popularity']=df_features['Movie_Popularity'].astype(str)
```

```python
X=df_features['Movie_Genre']+' '+df_features['Movie_Language']+' '+df_features['Movie_Popularity']+' '+df_features['Movie_Cast']+' '+df_features['Movie_Keywords']
```

```python
X
```

**GET FEATURES TEXT CONVERTS TO TOKEN-MEANS SPLITTING UP THE TEXT INTO SMALLER PARTS**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
model=TfidfVectorizer()
```

```python
X=model.fit_transform(X) #transforming the model X into tokens
```

```python
X
```

```python
X.shape
```

```python
print(X)
```

**USING COSINE SIMLARITY**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(X)
```

```python
similarity_score
```

```python
Favourite_movie=input("enter movie name :")
```

```python
all_titles=df['Movie_Title'].tolist()
```

```python
import difflib
```

```python
Movie_recommendation=difflib.get_close_matches(Favourite_movie,all_titles)
print(Movie_recommendation)
```

```python
close_match=Movie_recommendation[0]
print(close_match)
```

```python
index_of_close_match=df[df.Movie_Title==close_match]['Movie_ID'].values[0]
```

**GETTING A LIST OF SIMILAR MOVIES**

```python
recommendation_score=list(enumerate(similarity_score[index_of_close_match]))
#print(recommendation_score)
```

**SORTING MOVIES BASED ON THEIR RECOOMENDATION SCORE**

```python
sorted_movies=sorted(recommendation_score,key=lambda x:x[1],reverse=True)
#print(sorted_movies)
```

**PRINTING SIMILAR MOVIES BASED ON THEIR INDEX**

```python
print("top 50 movies for you:\n")
i=1
for movie in sorted_movies:
  index=movie[0]
  title_of_movie=df[df.index==index]['Movie_Title'].values[0]
  if(i<51):
    print(i,'.',title_of_movie)
    i=i+1
```

**top 10 recommendation system(only english movies right now)**

```python
favourite_movie=input("enter your favourite movie and based on that we recommend you movie :")
all_titles=df['Movie_Title'].tolist()
find_close_match=difflib.get_close_matches(favourite_movie,all_titles)
close_match=find_close_match[0]
index_movie=df[df.Movie_Title==close_match]['Movie_ID'].values[0]
recommendation_score=list(enumerate(similarity_score[index_movie]))
sorted_similar_movies=sorted(recommendation_score,key=lambda x:x[1],reverse=True)
print("TOP 10 MOVIES FOR YOU IS :/n")
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_movie=df[df.Movie_ID==index]['Movie_Title'].values
  if(i<11):
    print(i,'.',title_movie)
    i=i+1
```
